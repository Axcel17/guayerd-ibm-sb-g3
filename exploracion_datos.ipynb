### Insights basado en visualizaciones

```python
# Unir productos con detalle_ventas para obtener la categoría
df_productos_detalle = df_detalle_ventas.merge(df_productos[['id_producto', 'categoria']], on='id_producto')

# Unir con ventas para obtener la fecha
df_productos_ventas = df_productos_detalle.merge(df_ventas[['id_venta', 'fecha']], on='id_venta')

# Extraer el mes de la fecha
df_productos_ventas['mes'] = df_productos_ventas['fecha'].dt.month

# Calcular el monto por mes y categoría
monto_mes_categoria = df_productos_ventas.groupby(['mes', 'categoria'])['importe'].sum().reset_index()

# Crear pivot table para visualización
pivot_data = monto_mes_categoria.pivot(index='mes', columns='categoria', values='importe')

# Crear la visualización
plt.figure(figsize=(12, 6))
pivot_data.plot(kind='bar', ax=plt.gca())
plt.title('Monto de Ventas por Mes y Categoría', fontsize=14, fontweight='bold')
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Monto AR ($)', fontsize=12)
plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Calcular monto total por ciudad
df_ventas_clientes = df_ventas.merge(df_clientes[['id_cliente', 'ciudad']], on='id_cliente')
df_ventas_detalle = df_ventas_clientes.merge(df_detalle_ventas[['id_venta', 'importe']], on='id_venta')
monto_por_ciudad = df_ventas_detalle.groupby('ciudad')['importe'].sum()

# Crear figura con subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Proporción de clientes por ciudad
valores_prop = df_clientes['ciudad'].value_counts(normalize=True)
valores_prop.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_title('Proporción de Clientes por Ciudad', fontsize=14, fontweight='bold')
ax1.set_xlabel('Ciudad', fontsize=12)
ax1.set_ylabel('Proporción de Clientes', fontsize=12)
ax1.tick_params(axis='x', rotation=45, labelsize=10)
ax1.set_ylim(0, valores_prop.max() * 1.15)

# Agregar valores en las barras del primer gráfico
for i, v in enumerate(valores_prop):
    ax1.text(i, v + valores_prop.max() * 0.02, f'{v:.2%}', ha='center', va='bottom', fontweight='bold')

# Gráfico 2: Monto total por ciudad
monto_por_ciudad_ordenado = monto_por_ciudad.sort_values(ascending=False)
monto_por_ciudad_ordenado.plot(kind='bar', color='lightcoral', ax=ax2)
ax2.set_title('Monto Total de Ventas por Ciudad', fontsize=14, fontweight='bold')
ax2.set_xlabel('Ciudad', fontsize=12)
ax2.set_ylabel('Monto Total ($)', fontsize=12)
ax2.tick_params(axis='x', rotation=45, labelsize=10)

# Agregar valores en las barras del segundo gráfico
for i, v in enumerate(monto_por_ciudad_ordenado):
    ax2.text(i, v + monto_por_ciudad_ordenado.max() * 0.02, f'${v:,.0f}', 
             ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# 2.3 Visualización: Frecuencia de Métodos de Pago

plt.figure(figsize=(8, 5))
sns.countplot(data=df_ventas, x='medio_pago', 
              order=df_ventas['medio_pago'].value_counts().index, 
              palette='Set2')

plt.title('Métodos de Pago Más Usados', fontsize=12, fontweight='bold')
plt.xlabel('Método de Pago', fontsize=10)
plt.ylabel('Cantidad de Transacciones', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# Obtener las top 3 ciudades por importe
top_3_ciudades = monto_por_ciudad.head(3).index.tolist()
print("Top 3 ciudades por monto:", top_3_ciudades)

# Filtrar datos para las top 3 ciudades y calcular monto por método de pago
df_top_ciudades = df_ventas_detalle[df_ventas_detalle['ciudad'].isin(top_3_ciudades)]
monto_por_ciudad_medio_pago = df_top_ciudades.groupby(['ciudad', 'medio_pago'])['importe'].sum().reset_index()

# Crear pivot table para visualización
pivot_medios_pago = monto_por_ciudad_medio_pago.pivot(index='ciudad', columns='medio_pago', values='importe').fillna(0)

# Crear la visualización
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Barras apiladas por ciudad y método de pago
pivot_medios_pago.plot(kind='bar', ax=ax1, color=['lightblue', 'orange', 'lightgreen', 'pink'])
ax1.set_title('Distribución de Montos por Método de Pago\n(Top 3 Ciudades)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Ciudad', fontsize=12)
ax1.set_ylabel('Monto Total ($)', fontsize=12)
ax1.legend(title='Método de Pago', bbox_to_anchor=(1.05, 1), loc='upper left')
ax1.tick_params(axis='x', rotation=45)

# Gráfico 2: Porcentajes por ciudad
pivot_porcentajes = pivot_medios_pago.div(pivot_medios_pago.sum(axis=1), axis=0) * 100
pivot_porcentajes.plot(kind='bar', stacked=True, ax=ax2, color=['lightblue', 'orange', 'lightgreen', 'pink'])
ax2.set_title('Distribución Porcentual de Métodos de Pago\n(Top 3 Ciudades)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Ciudad', fontsize=12)
ax2.set_ylabel('Porcentaje (%)', fontsize=12)
ax2.legend(title='Método de Pago', bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Get top 3 cities by revenue
top_3_ciudades = monto_por_ciudad.head(3).index.tolist()
print("Top 3 ciudades por monto:", top_3_ciudades)

# Filter data for top 3 cities and calculate revenue by payment method
df_top_ciudades = df_ventas_detalle[df_ventas_detalle['ciudad'].isin(top_3_ciudades)]
# Merge to get product details sold in these cities
df_ventas_productos = df_top_ciudades.merge(df_detalle_ventas[['id_venta', 'id_producto', 'nombre_producto']], on='id_venta')

df_productos_ciudad = df_ventas_productos.groupby(['ciudad', 'nombre_producto'])['importe'].agg(['sum', 'count']).reset_index()
df_productos_ciudad.columns = ['ciudad', 'nombre_producto', 'importe_total', 'cantidad_transacciones']

# Get top 3 products by revenue in each city
top_productos_importe_ciudad = df_productos_ciudad.groupby('ciudad').apply(
    lambda x: x.nlargest(3, 'importe_total')[['nombre_producto', 'importe_total', 'cantidad_transacciones']]
).reset_index(level=0)

# Create improved visualization
fig, axes = plt.subplots(1, 3, figsize=(20, 8))
fig.suptitle('Top 3 Productos por Importe en las Top 3 Ciudades', 
             fontsize=16, fontweight='bold', y=0.95)

ciudades = top_3_ciudades
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']

# Enhanced charts by city
for i, ciudad in enumerate(ciudades):
    data_ciudad = top_productos_importe_ciudad[top_productos_importe_ciudad['ciudad'] == ciudad]
    
    # Use different colors for each product
    bars = axes[i].bar(range(len(data_ciudad)), data_ciudad['importe_total'], 
                      color=colors[:len(data_ciudad)], alpha=0.8, edgecolor='black', linewidth=0.8)
    
    axes[i].set_title(f'{ciudad}', fontsize=14, fontweight='bold', pad=15)
    axes[i].set_xlabel('Productos', fontsize=11)
    axes[i].set_ylabel('Importe Total ($)', fontsize=11)
    axes[i].set_xticks(range(len(data_ciudad)))
    
    # Truncate product names to 20 characters and wrap
    product_labels = []
    for prod in data_ciudad['nombre_producto']:
        if len(prod) > 20:
            # Split at 20 characters and add line break
            truncated = prod[:20] + '\n' + prod[20:40]
            if len(prod) > 40:
                truncated += '...'
        else:
            truncated = prod
        product_labels.append(truncated)
    
    axes[i].set_xticklabels(product_labels, rotation=0, ha='center', fontsize=9)
    
    # Add values on bars with better positioning
    for j, (v_importe, v_trans) in enumerate(zip(data_ciudad['importe_total'], data_ciudad['cantidad_transacciones'])):
        # Revenue value at top
        axes[i].text(j, v_importe + data_ciudad['importe_total'].max() * 0.03, f'${v_importe:,.0f}', 
                    ha='center', va='bottom', fontweight='bold', fontsize=10, color='darkblue')
        # Transaction count in center with better background
        axes[i].text(j, v_importe/2, f'{v_trans}\ntrans.', 
                    ha='center', va='center', fontweight='bold', fontsize=9, 
                    bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9, 
                             edgecolor='gray', linewidth=1))
    
    # Improve grid and formatting
    axes[i].grid(True, axis='y', alpha=0.3, linestyle='--')
    axes[i].set_ylim(0, data_ciudad['importe_total'].max() * 1.15)

plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()
```
